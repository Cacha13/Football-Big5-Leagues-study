```python
# Import of libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=pd.errors.SettingWithCopyWarning)
# Load CSV
players_df = pd.read_csv('C:\\Users\\ruiro\\OneDrive\\Ambiente de Trabalho\\Code\\premier league.csv', sep =';')
print(players_df.head())
```

La Liga Players Extraction

```python
# The leagues in the dataset
leagues = players_df.Comp.unique()
print(leagues)
# Create a la liga players Df
la_liga_players = players_df[players_df['Comp'] == 'La Liga']
la_liga_players.head()
```

```python
print(la_liga_players.columns)
list(la_liga_players.columns)
```

Selecting the columns to analyze

```python
la_liga_select = la_liga_players[['Player','Nation','Pos','Squad','Age','MP','Starts', 'Min','Goals']]
# Rename the columns names
la_liga_select.columns = ['Name', 'Nationality', 'Position', 'Team', 'Age', 'Total_Games', 'Start11', 'Total_Min', 'Goals']
la_liga_select.info()
```

```python
# add the columns MinPerGame and GoalPerGame 
la_liga_select['MinPerGame'] = (la_liga_select['Total_Min'] / la_liga_select['Total_Games']).astype(int)
la_liga_select['GoalPerGame'] = (la_liga_select['Goals'] / la_liga_select['Total_Games'])
la_liga_select.head()
```

```python
la_liga_select.describe()
```

```python
# Total Goals Scored
total_goals = la_liga_select.Goals.sum()
print(total_goals)
```

```python
unique_position = la_liga_select['Position'].unique()
print(unique_position)
```

Replace of Position 

```python
la_liga_select['Position'] = la_liga_select['Position'].replace({'MFFW': 'MF', 'MFDF': 'MF', 'FWMF': 'FW', 'FWDF' : 'FW', 'DFMF': 'DF', 'DFFW': 'DF'})
unique_position = la_liga_select['Position'].unique()
print(unique_position)
```

Pie Chart Goals vs Position

```python
plt.figure(figsize = (8,6))
goals_by_position = la_liga_select.groupby('Position')['Goals'].sum()
plt.pie(goals_by_position, autopct= '%.0f%%')
plt.legend(goals_by_position.index)
plt.title('Percentage of Goals per Position')
plt.axis('equal')
plt.show()
plt.clf()
```

Nationality Study

```python
# Number of Nations represented in La Liga 22/23
numb_nation = la_liga_select['Nationality'].nunique()
print(numb_nation)
```

```python
nation = la_liga_select.groupby('Nationality').size().sort_values(ascending = False)
nation_top10 = nation.head(10)
print(nation_top10)
# Barplot referring to the top10 nationalities
plt.figure(figsize = (8,6))
plt.bar(nation_top10.index, nation_top10.values, color = sns.color_palette('magma'))
plt.title('Top 10 Nations in La Liga 22/23')
plt.xlabel('Nationality')
plt.ylabel('Nº of players')
plt.show()
plt.clf()
```

```python
# The amount of nationalities for each team 
teams_nations = la_liga_select.groupby('Team')['Nationality'].nunique().sort_values(ascending = False)
print(teams_nations)
```

```python
# Barplot the teams_nations
plt.figure(figsize = (13,6))
plt.bar(teams_nations.index, teams_nations.values, color = sns.color_palette('viridis'))
plt.title('Nº of nationalities per team')
plt.xlabel('Team')
plt.ylabel('Nº of nationalities')
plt.xticks(rotation=60)
plt.show()
plt.clf()
```

```python
# spanish players per team 
teams_spain = la_liga_select[la_liga_select['Nationality'] == 'ESP'].groupby('Team').size().sort_values(ascending = False)
print(teams_spain)
# total players in each team
total_players_teams = la_liga_select.groupby('Team').size()
# proportion of spanish players per team
prop_spain_teams = (teams_spain / total_players_teams * 100).round().sort_values(ascending = False)
print(prop_spain_teams)
```

```python
# Plotting the Proportion of Spanish players per team
plt.figure(figsize=(13, 6))
plt.bar(prop_spain_teams.index, prop_spain_teams.values, color = sns.color_palette('viridis'))
plt.title(f'Proportion of Spanish Players per Team')
plt.xlabel('Team')
plt.ylabel(f'Proportion of Spanish Players (%)')
plt.xticks(rotation=60)
plt.show()
plt.clf()
```

Age Study

```python
# Youngest Player in La Liga 22/23
youngest_age = min(la_liga_select['Age'])
youngest_player = la_liga_select[la_liga_select['Age'] == youngest_age]
print(youngest_player['Age'])
print(youngest_player.Name + ', ' + youngest_player.Team)
```

```python
# Oldest player in LaLiga 22/23
oldest_age = max(la_liga_select['Age'])
oldest_player = la_liga_select[la_liga_select['Age'] == oldest_age]
print(oldest_player['Age'])
print(oldest_player.Name + ', ' + oldest_player.Team)
```

```python
# Players based on age 
under20 = la_liga_select[la_liga_select['Age'] < 20]
age20_25 = la_liga_select[(la_liga_select['Age'] >= 20) & (la_liga_select['Age'] <= 25)]
age25_30 = la_liga_select[(la_liga_select['Age'] > 25) & (la_liga_select['Age'] <= 30)]
age30_35 = la_liga_select[(la_liga_select['Age'] > 30) & (la_liga_select['Age'] <= 35)]
over35 = la_liga_select[la_liga_select['Age'] > 35]

x = np.array([under20['Name'].count(), age20_25['Name'].count(), age25_30['Name'].count(), age30_35['Name'].count(), over35['Name'].count()])
plt.pie(x, autopct = '%.0f%%')
plt.legend(['Under 20', '20 - 25', '26 - 30', '31 - 35', 'Over 35'])
plt.show()
plt.clf()
```

```python
# Total under20 players per team
under20_per_team = under20.groupby('Team').size().sort_values(ascending = False)
print(under20_per_team)
```

```python
# Total over35 players per team
over35_per_team = over35.groupby('Team').size().sort_values(ascending = False)
print(over35_per_team)
```

```python
# Average age in each team
average_age = la_liga_select.groupby('Team')['Age'].mean().round(2).sort_values(ascending = False)
print(average_age)
```

```python
# Boxplot 
plt.figure(figsize = (13,6))
sns.boxplot(x = 'Team', y = 'Age', data = la_liga_select, palette = 'rainbow')
plt.xticks(rotation = 60)
plt.show()
plt.clf()
```

Function with the Major Stats for a chosen Team in the Big5 Leagues

```python
# Creating a new df_players to be able to analyze also other leagues
players_selected = players_df[['Player','Nation','Pos','Squad','Age','MP','Starts', 'Min','Goals']]
# Rename the columns names
players_selected.columns = ['Name', 'Nationality', 'Position', 'Team', 'Age', 'Total_Games', 'Start11', 'Total_Min', 'Goals']
players_selected['MinPerGame'] = (players_selected['Total_Min'] / players_selected['Total_Games']).astype(int)
players_selected['GoalPerGame'] = (players_selected['Goals'] / players_selected['Total_Games']).round(2)
players_selected.info()
```

```python
def Stats(team):
    club = players_selected[players_selected['Team'] == team ]
    print(team + ' Total Goals in La Liga 22/23:')
    print(club['Goals'].sum())

    print(team + ' Average Age:')
    print(club['Age'].mean().round(2))

    print('')
    print(team + ' Top Scorer:')
    print(club[['Name', 'Goals']].sort_values(by = 'Goals', ascending = False).head(1))

    print('')
    print(team + ' Youngest Player:')
    younger_age = club.Age.min()
    young_player = club[club['Age'] == younger_age]
    print(young_player[['Name','Age']])

    print('')
    print(team + ' Oldest Player:')
    older_age = club.Age.max()
    old_player = club[club['Age'] == older_age]
    print(old_player[['Name','Age']])

    # Plots
    # Plotting the groups age in the team
    print('')
    print('')
    print('GRAPHICS:')
    print('')
    print('Percentage of Players in each age group:')

    under_20_squad = club[club['Age'] < 20]
    age20_25_squad = club[(club['Age'] >= 20) & (club['Age'] <= 25)]
    age25_30_squad = club[(club['Age'] > 25) & (club['Age'] <= 30)]
    age30_35_squad = club[(club['Age'] > 30) & (club['Age'] <= 35)]
    over35_squad = club[club['Age'] > 35]

    x = np.array([under_20_squad['Name'].count(), age20_25_squad['Name'].count(), age25_30_squad['Name'].count(), age30_35_squad['Name'].count(), over35_squad['Name'].count()])
    plt.pie(x, autopct = '%.0f%%')
    plt.legend(['Under 20', '20 - 25', '26 - 30', '31 - 35', 'Over 35'])
    plt.show()
    plt.clf()

    # Plotting the nationalities in the team:
    nation_squad = club.groupby('Nationality').size().sort_values(ascending = False)
    nation_squad_top10 = nation_squad.head(10)
    print(team + ' Top 10 nationalities')

    plt.figure(figsize = (8,6))
    plt.bar(nation_squad_top10.index, nation_squad_top10.values, color = sns.color_palette('magma'))
    plt.title('Top 10 Nations in ' + team + ':')
    plt.xlabel('Nationality')
    plt.ylabel('Nº of players')
    plt.show()
    plt.clf()

    # Plot the best scorers and the minutes per goal
    top5_scorers = club.sort_values(by = 'Goals', ascending = False).head(5)
    top5_scorers['Min_to_score'] = (top5_scorers['Total_Min']/ top5_scorers['Goals']).astype(int)
    
    plt.figure(figsize = (14,6))
    # Bar plot for Goals
    plt.subplot(1, 2, 1)
    plt.bar(top5_scorers['Name'], top5_scorers['Goals'], color='skyblue')
    plt.xticks(rotation=75)
    plt.title(team + " Top 5 Scorers - Goals")
    plt.ylabel("Goals")

    # Bar plot for minutes to score
    plt.subplot(1, 2, 2)
    plt.bar(top5_scorers['Name'], top5_scorers['Min_to_score'], color='salmon')
    plt.xticks(rotation=75)
    plt.title("Top 5 Scorers - Minutes to score a goal")
    plt.ylabel("Minutes")
```

```python
Team = 'Liverpool'
Stats(Team)
```
